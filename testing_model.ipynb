```python
import cv2
import numpy as np
from imutils import paths
from utils.model import NN
from matplotlib import pyplot as plt

test_images_folder = 'captcha_test' 
trained_model_location = 'trained_model'

captcha_image_files = list(paths.list_images(test_images_folder))
captcha_image_files = np.random.choice(captcha_image_files, size=(10,), replace=False)

model = NN.load(trained_model_location)

for image_file in captcha_image_files:
    image = cv2.imread(image_file)
    captcha_text = model.solve(image)
    print("CAPTCHA text is: {}".format(captcha_text))
    plt.imshow(image)
    plt.show()
```
